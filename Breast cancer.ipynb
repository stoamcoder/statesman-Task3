```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
from pandas import DataFrame as df
```

```python
cancer = pd.read_csv('breastcancer.csv', index_col='id')
cancer.head()
```

```python
cancer.describe()
```

```python
#slicing out the continuous variable

#cancer_dimension = cancer[["radius_mean", "texture_mean","perimeter_mean", "area_mean"]]
cancer_dimension = cancer[["radius_mean", "texture_mean"]]
cancer_morphology = cancer[["smoothness_mean", "compactness_mean", "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean"]]
#cancer_dimension.tail()
#sns.heatmap(data=cancer_dimension, annot=False)
sns.heatmap(data=cancer_dimension, annot=False)
```

```python
#checking the distribution of malignant vs benign in the breast cancer data
barc = cancer.diagnosis
barc
```

```python
#plotting the distribution
plt.hist(barc)
plt.ylabel("Number of diagnostics")
```

```python
sorted_cancer_area = cancer.sort_values(by = ['area_mean'], ascending=True)
sorted_cancer_area.head()
sorted_cancer_area.tail()
df(sorted_cancer_area)
#sns.distplot(a=cancer['radius_mean'], kde=True)
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x="area_mean", y="concavity_mean", hue="diagnosis", data=cancer)

sns.lmplot(x="area_mean", y="concave points_mean", hue="diagnosis", data=cancer)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x="area_mean", y="radius_mean", hue="diagnosis", data=cancer)
sns.lmplot(x="area_mean", y="symmetry_mean", hue="diagnosis", data=cancer)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x="area_mean", y="perimeter_mean", hue="diagnosis", data=cancer)
sns.lmplot(x="area_mean", y="compactness_mean", hue="diagnosis", data=cancer)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x="area_mean", y="fractal_dimension_mean", hue="diagnosis", data=cancer)
sns.lmplot(x="area_mean", y="perimeter_mean", hue="diagnosis", data=cancer)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lmplot(x="area_mean", y="texture_mean", hue="diagnosis", data=cancer)
sns.lmplot(x="area_mean", y="compactness_mean", hue="diagnosis", data=cancer)
sns.lmplot(x="area_mean", y="smoothness_mean", hue="diagnosis", data=cancer)
plt.show()
```

```python
plt.figure(figsize=(14,5))
sns.swarmplot(x=cancer['area_mean'], y=cancer['concave points_mean'], hue=cancer['diagnosis'])
plt.show()
```

```python
sns.jointplot(x=cancer['compactness_mean'], y=cancer['area_mean'], kind="kde")
sns.jointplot(x=cancer['compactness_mean'], y=cancer['concave points_mean'], kind="kde")
```

```python
sns.regplot(x=cancer['concave points_mean'], y=cancer['area_mean'])
plt.xlabel("No of concave points detected") 
plt.ylabel("Area of the tumor")
```
